## Training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
iris.data.shape, iris.target.shape
```

```python
training_data, testing_data, training_labels, testing_labels = train_test_split(
    iris.data,
    iris.target,
    test_size=0.4,
    random_state=0
)
```

```python
training_data.shape, training_labels.shape
```

```python
testing_data.shape, testing_labels.shape
```

## Build the model

```python
from sklearn import svm
```

```python
model = svm.SVC(C=1, kernel='linear')
```

```python
print(model)
```

## Training the model

```python
model.fit(training_data, training_labels)
```

## Testing the model

```python
from sklearn.metrics import accuracy_score
```

```python
test_data_predictions = model.predict(testing_data)
print (test_data_predictions)
```

```python
print (testing_labels)
```

```python
accuracy_score(test_data_predictions, testing_labels)
```
